In this notebook, we'll simulate the spread of an infectious disease using the SIR model. We'll define the model, run simulations, and visualize the results.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from matplotlib.animation import FuncAnimation
from IPython.display import Image, display
from ipywidgets import interact, FloatSlider
```

```python
class disease:
    def __init__(self, name, beta, gamma):
        self.name = name
        self.beta = beta
        self.gamma = gamma

# Create a list of Disease objects
diseases = [
    disease("Influenza", 0.18, 0.014),
    disease("Measles", 1.8, 0.007),
    disease("SARS-CoV-2", 0.2, 0.01),
    disease("Ebola", 0.25, 0.004),
    disease("Rhinovirus", 0.2, 0.014)
]
```

```python
# Define the SIR model differential equations with seasonal effects
def sir_model(y, t, beta, gamma):
    S, I, R = y
    # Seasonal effect varies sinusoidally over the year
    beta_seasonal = beta * (1 + 0.5 * np.sin(2 * np.pi * t / 365))
    dS_dt = -beta_seasonal * S * I
    dI_dt = beta_seasonal * S * I - gamma * I
    dR_dt = gamma * I
    return [dS_dt, dI_dt, dR_dt]

# Function to run the simulation with vaccination
def run_simulation(S0, I0, R0, disease, days, vaccination_rate):
    if vaccination_rate > 0:
        vaccinated = S0 * vaccination_rate
        S0 -= vaccinated
        R0 += vaccinated
    
    y0 = [S0, I0, R0]
    t = np.linspace(0, days, days)
    solution = odeint(sir_model, y0, t, args=(disease.beta, disease.gamma))
    return t, solution.T

# Enhanced visualization with animation
def animate_simulation(t, S, I, R, disease, vaccination_rate):
    fig, ax = plt.subplots(figsize=(14, 10))
    ax.set_xlim(0, max(t))
    ax.set_ylim(0, 1)
    ln1, = ax.plot([], [], 'blue', label='Susceptible')
    ln2, = ax.plot([], [], 'red', label='Infected')
    ln3, = ax.plot([], [], 'green', label='Recovered')

    def init():
        ln1.set_data([], [])
        ln2.set_data([], [])
        ln3.set_data([], [])
        return ln1, ln2, ln3

    def update(frame):
        ln1.set_data(t[:frame], S[:frame])
        ln2.set_data(t[:frame], I[:frame])
        ln3.set_data(t[:frame], R[:frame])
        return ln1, ln2, ln3

    ax.set_xlabel('Time (days)', fontsize=14)
    ax.set_ylabel('Proportion of Population', fontsize=14)
    ax.set_title(f'Animated SIR Model for {disease.name}\nBeta: {disease.beta}, Gamma: {disease.gamma}, Vaccination Rate: {vaccination_rate}', fontsize=16)
    ax.legend()

    anim = FuncAnimation(fig, update, frames=len(t), init_func=init, blit=True, repeat=False)

    # Save animation to a local path
    anim_path = 'sir_animation.gif'
    anim.save(anim_path, writer='pillow', fps=30)
    
    # Display the GIF using IPython's Image class
    return Image(filename=anim_path)

# Initial conditions and parameters
S0 = 0.99  # Initial susceptible population (99%)
I0 = 0.01  # Initial infected population (1%)
R0 = 0.0   # Initial recovered population (0%)

beta = 0.3   # Infection rate
gamma = 0.1  # Recovery rate
days = 365   # Number of days to simulate

# Run simulations for different scenarios
vaccination_rates = [0.1, 0.3, 0.4]

for vaccination_rate in vaccination_rates:
    for disease in diseases:
        t, (S, I, R) = run_simulation(S0, I0, R0, disease, days, vaccination_rate)
        display(animate_simulation(t, S, I, R, disease, vaccination_rate))
        plt.figure(figsize=(10, 7))
        plt.plot(S, I, label='Susceptible vs Infected', color='blue')
        plt.xlabel('Susceptible Population')
        plt.ylabel('Infected Population')
        plt.title('Phase Portrait of the SIR Model')
        plt.legend()
        plt.grid(True)
        plt.show()
        plt.figure(figsize=(10, 7))
        plt.imshow([I], aspect='auto', cmap='hot', origin='lower', extent=[0, len(I), 0, 1])
        plt.colorbar(label='Infected Population')
        plt.xlabel('Time (days)')
        plt.ylabel('Proportion of Population')
        plt.title('Heatmap of Infection Over Time')
        plt.show()
```

```python
def sir_model_complex(y, t, beta, gamma, birth_rate, death_rate):
    S, I, R = y
    beta_seasonal = beta * (1 + 0.5 * np.sin(2 * np.pi * t / 365))
    dS_dt = birth_rate - beta_seasonal * S * I - death_rate * S
    dI_dt = beta_seasonal * S * I - gamma * I - death_rate * I
    dR_dt = gamma * I - death_rate * R
    return [dS_dt, dI_dt, dR_dt]

def run_simulation_complex(S0, I0, R0, disease, days, birth_rate, death_rate, vaccination_rate):
    if vaccination_rate > 0:
        vaccinated = S0 * vaccination_rate
        S0 -= vaccinated
        R0 += vaccinated
    
    y0 = [S0, I0, R0]
    t = np.linspace(0, days, days)
    solution = odeint(sir_model_complex, y0, t, args=(disease.beta, disease.gamma, birth_rate, death_rate))
    return t, solution.T

# Enhanced visualization with animation
def animate_simulation_complex(t, S, I, R, disease, birth_rate, death_rate, vaccination_rate):
    fig, ax = plt.subplots(figsize=(14, 10))
    ax.set_xlim(0, max(t))
    ax.set_ylim(0, 1)
    ln1, = ax.plot([], [], 'blue', label='Susceptible')
    ln2, = ax.plot([], [], 'red', label='Infected')
    ln3, = ax.plot([], [], 'green', label='Recovered')

    def init():
        ln1.set_data([], [])
        ln2.set_data([], [])
        ln3.set_data([], [])
        return ln1, ln2, ln3

    def update(frame):
        ln1.set_data(t[:frame], S[:frame])
        ln2.set_data(t[:frame], I[:frame])
        ln3.set_data(t[:frame], R[:frame])
        return ln1, ln2, ln3

    ax.set_xlabel('Time (days)', fontsize=14)
    ax.set_ylabel('Proportion of Population', fontsize=14)
    ax.set_title(f'Animated SIR Model for {disease.name}\nBeta: {disease.beta}, Gamma: {disease.gamma}, Birth: {birth_rate}, Death: {death_rate}, Vaccination Rate: {vaccination_rate}', fontsize=16)
    ax.legend()

    anim = FuncAnimation(fig, update, frames=len(t), init_func=init, blit=True, repeat=False)

    # Save animation to a local path
    anim_path = 'sir_animation.gif'
    
    # Display the GIF using IPython's Image class
    return Image(filename=anim_path)

# Run simulations for different scenarios
vaccination_rates = [0.1, 0.3, 0.4]
birth_rate = 1.85
death_rate = 0.76

for vaccination_rate in vaccination_rates:
    for disease in diseases:
        t, (S, I, R) = run_simulation_complex(S0, I0, R0, disease, days, birth_rate, death_rate, vaccination_rate)
        display(animate_simulation_complex(t, S, I, R, disease, birth_rate, death_rate, vaccination_rate))
```
